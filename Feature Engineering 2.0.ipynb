```python
def cleaned_mean(arr):

    if arr==[None]:
        m=0
    else:
        print(arr)
        arr = list(filter(lambda x: x!= ['ooa'], arr))
        arr = list(filter(lambda x: x!= [], arr))
        df = pd.DataFrame(arr,columns=['value'])
        print(df)
        max_treshold=df['value'].quantile(0.80)
        df=df[df['value']<max_treshold]
        df=df[df['value']>0]
        m=np.mean(df.value)
    return m
```

```python
def remove_ooa(arr,m):
    count=0
    array=[]
    if arr==[None]:
        array=[None]
    else:
        for i in range(len(arr)):
            
            if type(arr[i]) == type([]):
                if arr[i]==[]:
                    array.append(m)
                elif arr[i]==['ooa']:
                    count+=1
                    array.append(m)
            else:
                array.append(arr[i])
    return array,count
            
```

```python
def clean(x):
    mod_x=[]

    for j in x:
   
        mn=cleaned_mean(j)
        s,_=remove_ooa(j,mn)
        mod_x.append(s)
    return mod_x
```

```python
new_sT=clean(f_Ts)
new_sT
```

```python
def reliability_matrix(arr):
    rel_matrix=[]

    for i in range(len(arr)):
        if arr[i]==[None]:
            s=[None]
        else:
            df = pd.DataFrame(arr[i],columns=['value'])
            print(arr[i]) 
            outliers = setup(df ,silent=True,numeric_features = ["value"])
            knn = create_model('knn')
            knn_results = assign_model(knn)
            s=np.array(knn_results.Anomaly_Score)
            s=1-s
            s=np.round(100*s,2)
            s=[0 if x==100 else x for x in s]
        rel_matrix.append(s)
    return rel_matrix
```

```python
rel_matrix=reliability_matrix(new_sT)
```

```python
def shape(a):
    if not type(a) == list:
        return []
    return [len(a)] + shape(a[0])
```

```python
def loop(s,r):
    dim=len(shape(r))
    if dim==1:
        for i in range(len(r)):
            if not r[i]==None:
                if r[i]<75:
                    s[i]=s[i-1]
    else:
        for j in range(len(s)):
            loop(s[j],r[j])
        
    return s
            
        
```

```python
filtered_sT=loop(new_sT,rel_matrix) 
filtered_sT
```
