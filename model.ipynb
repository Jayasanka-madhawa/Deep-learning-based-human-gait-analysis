```python
import numpy as np
import pandas as pd
import pprint, pickle
```

```python
amani_file = open('feature_vector/FV_side_amani.pkl', 'rb')
gayanjali_file = open('feature_vector/FV_side_gayanjali.pkl', 'rb')
harshi_file = open('feature_vector/FV_side_harshi.pkl', 'rb')
hasitha_file = open('feature_vector/FV_side_hasitha.pkl', 'rb')
isurika_file = open('feature_vector/FV_side_isurika.pkl', 'rb')

data1 = pickle.load(amani_file)
data2 = pickle.load(gayanjali_file)
data3 = pickle.load(harshi_file)
data4 = pickle.load(hasitha_file)
data5 = pickle.load(isurika_file)
#pprint.pprint(sT)

amani_file.close()
gayanjali_file.close()
harshi_file.close()
hasitha_file.close()
isurika_file.close()
```

```python
data1
```

```python
data=[data1,data2,data3,data4,data5]
```

```python
type(data[0])
```

```python
maxLen = max(map(lambda x:len(x[0]), data))
maxLen

data_arr=[]
c=0
for k in data:
    c+=1
    s=[]
    d=maxLen//len(k[0])
    r=maxLen%len(k[0])
    for i in k:
        i=list(i)
        arr=i*d+i[:r]
        s.append(arr)
    s=np.array(s)
    s=s.T
    data_arr.append(s)
data_arr=np.array(data_arr)
data_arr.shape
```

```python
data_arr[0].shape[0]
```

```python
s=np.full(shape=data_arr[0].shape[0],fill_value=0)
s=s.reshape(-1,1)
s
```

```python
final_data=[]
for i in range(len(data_arr)):
    s=np.full(shape=data_arr[i].shape[0],fill_value=i)
    s=s.reshape(-1,1)
    array = np.append(data_arr[i],s, axis = 1)
    if final_data==[]:
        final_data=array
    else:
        final_data=np.append(final_data,array,axis=0)
print(final_data.shape)
    
```

```python
df=pd.DataFrame(final_data,columns=['p1','p2','p3','p4','p5','p6','target'])
df
```

```python
df_input=df.drop(['target'], axis=1)
df_input
```

```python
df_input.mean()
```

```python

df_input.fillna(method='ffill')
```

```python
m=df_input.mean()
values = {'p1':m[0],'p2':m[1],'p3':m[2],'p4':m[3],'p5':m[4],'p6':m[5]}
```

```python
df_input=df_input.fillna(value=values)
df_input
```

```python
df.target
```

```python

import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
%matplotlib inline
```

```python

X_train, X_test, y_train, y_test = train_test_split(df_input,df.target,test_size=0.2, random_state=25)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
X_train
```

```python
model = keras.Sequential([
    keras.layers.Dense(5, input_shape=(6,), activation='sigmoid')
    
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(X_train, y_train, epochs=1000)
```

```python
y_predicted = model.predict(X_test)
y_predicted[1]
```

```python
y_test
```

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(6,)),
    keras.layers.Dense(100, activation='relu'),
    keras.layers.Dense(5, activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(X_train, y_train, epochs=1000)
```

```python
model.evaluate(X_test, y_test)
```
